<h3> Model Fitting <h3>

```python
#Feature Imporatance with decisionTree

train_data=pd.read_csv("cleaned_final.csv")
Y=train_data.blueWins
features=train_data.drop(['blueWins'],axis=1)
features

model=DecisionTreeClassifier()
model.fit(features,Y)
importance = model.feature_importances_

plt.bar([x for x in range(len(importance))], importance)
plt.show()
plt.savefig('decisionTree_feature_importance.png')

for i,v in enumerate(importance):
  print('Feature: %0d, Score: %.5f' % (i,v))


```

```python
#Feature Imporatance with LogisticRegression

model2=LogisticRegression()
model2.fit(features,Y)
importance2 = model.feature_importances_

plt.bar([x for x in range(len(importance2))], importance)
plt.show()
plt.savefig('Logistic_feature_importance.png')

for i,v in enumerate(importance2):
	print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
#spliting data

X_train,X_test,Y_train,Y_test=train_test_split(features,Y,test_size=0.2,random_state=0)

#Fitting DecisionTree & estimating accuracy

decisionTree=model.fit(X_train,Y_train)
Y_pred = decisionTree.predict(X_test)
score1 = accuracy_score(Y_test, Y_pred)
print("DecisionTree Accuracy : ", score1)

#Fitting LogisticRegression & estimating accuracy

logisticReg=model2.fit(X_train,Y_train)
Y_pred1=logisticReg.predict(X_test)
score2 = accuracy_score(Y_test, Y_pred1)
print("Logistic Regression Accuracy : ", score2)


#Fitting ElasticNet & estimating accuracy

regr = ElasticNet(alpha=0.01)
regr.fit(X_train,Y_train)

Y_pred2= regr.predict(X_test)
Y_pred2_2=np.round(np.clip(Y_pred2, 0, 1))
score3 = accuracy_score(Y_test, Y_pred2_2)
cm_elasticNet=confusion_matrix(Y_test, Y_pred2_2)

print("ElasticNet Accuracy : ", score3)
print("ElasticNet Confusion matrix : ", cm_elasticNet)

#Fitting Ridget & estimating accuracy

regr2 = Ridge(alpha=0.01)
regr2.fit(X_train,Y_train)

Y_pred3= regr2.predict(X_test)
Y_pred3_2=np.round(np.clip(Y_pred3, 0, 1))
score4 = accuracy_score(Y_test, Y_pred3_2)
cm_ridge=confusion_matrix(Y_test, Y_pred3_2)

print("Ridge Accuracy : ", score4)
print("Ridge Confusion matrix : ", cm_ridge)

#Fitting Lasso & estimating accuracy

regr3= Lasso(alpha=1.0)
regr3.fit(X_train,Y_train)

Y_pred4= regr3.predict(X_test)
Y_pred4_2=np.round(np.clip(Y_pred4, 0, 1))
score5 = accuracy_score(Y_test, Y_pred4_2)
cm_lasso=confusion_matrix(Y_test, Y_pred4_2)


print("Lasso Accuracy : ", score5)
print("Lasso Confusion matrix : ", cm_lasso)
```
